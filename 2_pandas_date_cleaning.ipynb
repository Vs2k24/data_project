```python
#import all the neccessary libraries
import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset

#load the data 
dataset=load_dataset('lukebarousse/data_jobs')
#convert it into a dataframe 
df=dataset['train'].to_pandas()

#data cleanup 
df['job_posted_date']=pd.to_datetime(df['job_posted_date'])
```

```python
#now we want to access  the first 10 rows of 2 columns
#df.loc[:10,'col1':'col2']
df.loc[:10,'salary_year_avg':'salary_hour_avg']
```

```python
#now lets calculate the median values for the salary_year_avg
#df['salary_year_avg'] means that we are accessing the salary_year_avg col from the dataframe
median_salary_year=df['salary_year_avg'].median()
```

```python
median_salary_hour=df['salary_hour_avg'].median()
```

```python
#now we are going to create a new dataframe and set it equal to the original dataframe we had
df_filled=df
#now we will fill all the empty places or the na places with the median_salary_year value
#so in the new dataframe we will access the salary_year_avg col and wherever u find empty or na just fill in the median_year_salary value
df_filled['salary_year_avg'].fillna(median_salary_year)
```

```python
#but theres a problem here this only put the median_salary_year values in the empty place but its not updated
#if we try to see the column well see the values have still not yet been updated
df_filled['salary_year_avg']
```

```python
#so inorder to fix this we reassign it to the same col
df_filled['salary_year_avg']=df_filled['salary_year_avg'].fillna(median_salary_year)
```

```python
df_filled['salary_year_avg']
```

```python
#similarly for the salary_hour_avg
df_filled['salary_hour_avg']=df_filled['salary_hour_avg'].fillna(median_salary_hour)
```

```python
df_filled['salary_hour_avg']
```

```python
#now we can see the first 10 rows value of the 2 cols salary_year_avg and salary_hour_avg and this time there wont be any null or na values
df_filled.loc[:10,'salary_year_avg':'salary_hour_avg']
```

```python
#so in the above step what we did was removing na values by .dropna(subset='col') method
#then we found out the median df[colname].median()
#then we created a new dataframe which is set to the original df : df_filled=df
#now we filled all the empty or na places with the help of fillna(the value u want)
#in order to see the updated values u need to reassign to the same col of the dataframe
#in order to access the first 10 rows of a particular col use df.loc[0:10,'colname']
```

now we will see how to remove duplicates froma dataframe

```python
#we wont touch the original dataframe df therefore we create another dataframe and assign it to be equal to the orignal df and make changes in that
#lets create another df and make it equal to the df_filled dataframe
df_unique=df_filled
#now we will drop the duplicate rows
df_unique.drop_duplicates()
#to save the changes assign it to the same df
df_unique=df_unique.drop_duplicates()
df_unique
```

```python
#now we dont know how many duplicate rows did we drop
#to see that
print('length of original df : ' , len(df_filled))
print('length of drop duplicates df (df_unique) : ' , len(df_unique))
print('rows dropped : ' , len(df_filled)-len(df_unique))
```

```python
#now we are going to drop the duplicates in a particular col that we want by drop_dupliactes(subset=['colname'])
df_unique=df_unique.drop_duplicates(subset=['job_title','company_name'])
print('length of original df : ' , len(df_filled))
print('length of drop duplicates df (df_unique) : ' , len(df_unique))
print('rows dropped : ' , len(df_filled)-len(df_unique))
```

copy() method

```python
#loading the dataset
dataset=load_dataset('lukebarousse/data_jobs')
#we are creating another df and converting our original dataset into dataframe
df_original=dataset['train'].to_pandas()
```

```python
df_original
```

```python
#copy creates a copy of the dataframe
#in the above we first loaded the dataset and then 
#converted dataset into df and stored in a new dataframe
#so we did this way because if we make changes in our new df that we created
#it wont affect the original dataset whenever we modify our new dataframe
#theres another way to create a df
#df_new=df_original (new df is equal to the original dataset that we create)
#but the problem in this is that whenver we try to make changes or modify our new df
#our original df will also get affected by it and the values might get updated
#we see that id(dat_new)=id(data_original) they are referencing the same df thats why theres a problem
#therefore do:
#dataset=load_dataset('pathname')
#df_new=dataset['train'].to_pandas()
```

```python
#so instead of doing df_new=df_original what well first do is:
#load the dataset  and convert the dataset into df and store in the new df
#now again create a new df and use the copy method to store the value
df_new=df_original.copy()
```

```python
print('id of the df_original:   ', id(df_original))
print('id of the df_new:     ', id(df_new))
print('are the two df same:      ',  id(df_original)==id(df_new))
```

```python
#now we see that both the ids are different 
#so if u make changes in one df it wont affect the original df
```
