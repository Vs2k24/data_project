```python
#import neccessary libraries
import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset
import ast

#load the dataset
dataset=load_dataset('lukebarousse/data_jobs')
df=dataset['train'].to_pandas()

#data cleanup
df['job_posted_date']=pd.to_datetime(df['job_posted_date'])
df['job_skills']=df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
#now create a df that only contains data analyst jobs in the us
df_DA_US=df[(df['job_title_short']=='Data Analyst') & (df['job_country']=='United States')].copy()
df_DA_US
```

```python
df_DA_US['salary_year_avg'].plot(kind='hist',bins=30,edgecolor='black')
#we can make our x axis little bit more better
#xlim(lower,till where u want to go)
plt.xlim(0,250000)
#we will also format the x axis
#first get the current axis
ax=plt.gca()

#we want to make our x axis values(labels) look better
ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'${int(x/1000)}K'))

plt.title('distribution of united states data analyst yearly salaries')
plt.xlabel('yearly salary')
plt.ylabel('no of jobs')
```
