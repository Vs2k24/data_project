for index management we use the sort_index() method

```python
#import all the neccessary libraries
import pandas as pd 
from datasets import load_dataset
import matplotlib.pyplot as plt

#load the data
dataset=load_dataset('lukebarousse/data_jobs')
#convert datatset into dataframe
df=dataset['train'].to_pandas()

#data cleanup
df['job_posted_date']=pd.to_datetime(df['job_posted_date'])

```

```python
#df.index displays the current index of the dataframe
df.index
```

```python
#df.index.name here we are naming the index by job_index
''' 
before naming our data table was like this:
      Name  Age    City
0    Alice   25     NYC
1      Bob   30  London
2  Charlie   35   Tokyo
->here we can see that the index col(0,1,2) does not have a header
after writing index.name we can give a name to the column
then our table will look like this:

          Name  Age    City
job_index                  
0        Alice   25     NYC
1          Bob   30  London
2      Charlie   35   Tokyo

->here we see that the index column now has a header or a heading or a label with the name job_index
'''
df.index.name='job_index'
```

three methods to focus on:
df.reset_index()
df.set_index()
df.sort_index()

reset_index()

```python
#lets create a new df that its filtered for jobs particulary for the united states
#lets create a new df which oly has those jobs which are present in the us
#so give me the jobs that are in the us only
#It filters your DataFrame to keep only the job postings from the United States.
''' 
df[df['job_country'] == 'United States'] with this specific syntax.:
Think of it as asking your DataFrame: "Hey, show me all the rows WHERE job_country equals 'United States'"
its like telling the main df that from the df we are accessing the job_country and checking if country is us or not

Part 1: df['job_country'] == 'United States'
This creates a TRUE/FALSE mask for each row

Example output: [True, False, True, False, True, ...]

Each True means "this row is from United States"

Each False means "this row is from another country"

Part 2: df[ ] (The Filter)
When you put the TRUE/FALSE list inside df[ ]

Pandas says: "Give me only the rows where you have TRUE"
'''
df_usa=df[df['job_country']=='United States']
df_usa
```

```python
#here we see that in job index column: some indexes are missing means after 0 we directly got 3 (1,2 are missing) then after 3 we directly got 5 (4 index is missing) and so on
#so what we do here is we apply reset_index()
#"Take the row labels from df_usa, make them a new column, give it a new simple index, and save that change directly to df_usa."
#what happened here was we used reset_index to craete a new column and gave it a simple index like 0,1,2,3,4...
df_usa.reset_index(inplace=True)
df_usa
```

set_index()

```python
#set_index: if we dont want the boriing simple index like 0,1,2,3,4.. like we did in reset_index
#we can just set index that means we used the job index columns ki value and made it our index
#job index =>0,3,5.. and now our index also became 0,3,5..
#inplace = true makes the changes permanent in the original frame too
df_usa.set_index('job_index',inplace=True)
df_usa
```

```python
#reset_index=> makes a new column and gives simple indexing 0,1,2,3..
#set_index=> replaces the boring index with the columns value we want
```

sort_index()

```python
#here we are creating a summary table in which the rows display the job_title_short and the value that we are working on is the salary_year_avg
#and in the value we are applying three functions: median, min, max
median_pivot=df_usa.pivot_table(values='salary_year_avg',index='job_title_short',aggfunc=['median','min','max'])
median_pivot
```

```python
#we can sort this by median values
#"Sort this table by the 'salary_year_avg' column that's inside the 'median' section."
#It will rearrange the rows from lowest to highest median salary
median_pivot.sort_values(by=('median','salary_year_avg'))
```

```python
#u can either use sort_values(by=(method,colname))
#or sort_index()
median_pivot.sort_index(inplace=True)
```
