```python
#import all the necessary libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

#load the dataset
dataset=load_dataset('lukebarousse/data_jobs')
#convert dataset into df
df=dataset['train'].to_pandas()

#data cleanup
df['job_posted_date']=pd.to_datetime(df['job_posted_date'])
```

```python
#we are trying to merge to df one of data jobs and the other of software jobs
#data jobs df we already have (5_exercise)
df_usa=df[df['job_country']=='United States'].copy()
df_usa
```

```python
df_usa['job_posted_month']=df['job_posted_date'].dt.strftime('%B')
df_usa
```

```python
df_usa_pivot=df_usa.pivot_table(index='job_posted_month',columns='job_title_short',aggfunc='size')
df_usa_pivot
```

```python
df_usa_pivot.reset_index(inplace=True)
df_usa_pivot['month_no']=pd.to_datetime(df_usa_pivot['job_posted_month'],format='%B').dt.month
df_usa_pivot
```

```python
df_usa_pivot.sort_values('month_no',inplace=True)
```

```python
df_usa_pivot.set_index('job_posted_month',inplace=True)
df_usa_pivot
```

```python
df_usa_pivot.drop(columns='month_no',inplace=True)
df_usa_pivot
```

```python
#so the above table is our data_jobs summary table 
```

```python
#now for software jobs
pd.read_csv("https://lukeb.co/software_csv")
```

```python
#here we see that theres a index but in our data jobs df we had job_posted_month in our index
#so what we can do is set_index() or:
df_usa_software_pivot=pd.read_csv("https://lukeb.co/software_csv", index_col='job_posted_month')
df_usa_software_pivot
```

```python
#now we will merge these 2 dataframes
```

```python
#so we will write data_jobs df and merge software_jobs df and for both of these we are using index job_posted_month
#here on means the common key or the connection point
df_usa_merged=df_usa_pivot.merge(df_usa_software_pivot, on="job_posted_month")
```

```python
#so we have combined both the df
```

```python
''' explaination of on in df_usa_pivot.merge(df_usa_software_pivot, on='job_posted_month')
Imagine you have two different pages in an address book:

Page A is sorted by Name and has columns: Name and Phone Number.

Page B is sorted by Name and has columns: Name and Email Address.

If you want to combine them, you look for the common value: the Name. You find "John Smith" on Page A and then find "John Smith" on Page B, and then you write his phone number and email on the same line in your new list.

In this case, on='job_posted_month' is doing the exact same thing.

It finds a row in the first table (df_usa_pivot) where the job_posted_month is, for example, 2024-01.

It then looks into the second table (df_usa_software_pivot) for the row that also has job_posted_month equal to 2024-01.

It then takes the rest of the columns from both of those matching rows and puts them together into a single row in the new, merged table.
'''
```

```python
#now we want the top_5 jobs 
#so we will create a list of the top_5 jobs
#.sum() will sum all the columns value
#Go through each column in this report card and add up ALL the numbers from top to bottom. Give me the grand total for each category.
top_5=df_usa_merged.sum()
top_5
```

```python
#we want it to be in descending order and we only need the top 5 values
top_5=df_usa_merged.sum().sort_values(ascending=False).head(5)
top_5
```

```python
#we need the index and then convert index to list
top_5=df_usa_merged.sum().sort_values(ascending=False).head(5).index.to_list()
top_5
```

```python
df_usa_merged[top_5].plot(kind='line')
plt.title('monthly job postings for the top tech jobs in the us')
plt.xlabel('2023')
plt.ylabel('job_count')
plt.ylim(0,20000)
plt.legend()
plt.show()
```
