apply() - calculate project salary next year

```python
#import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset

#load dataset
dataset=load_dataset('lukebarousse/data_jobs')
df=dataset['train'].to_pandas()

#data cleanup
df['job_posted_date']=pd.to_datetime(df['job_posted_date'])
```

```python
df['salary_year_avg']
```

```python
#Gets only the non-empty salary values from your job data
#in my df im applying pd.notna to get only the not empty values
#by accessing the salary_year_avg col of my df and after filtering show only the salary_year_avg column with non empty values
#df[pd.notna(df['salary_year_avg'])]['salary_year_avg']
#df[pd.notna(df['salary_year_avg'])]['salary_year_avg']
df_salary=df[pd.notna(df['salary_year_avg'])]
df_salary
```

```python
#since we want to find the projected year salary for the next year well try to find the inflation
#lets say for next year the inflation is salary*1.03
#therefore lets create a function to find the inflation
def projected_salary(salary):
    return salary * 1.03

#now we access the salary_year_avg col from our df_salary dataframe and apply the function to all the salaries
df_salary['salary_year_avg'].apply(projected_salary)

```

```python
#lets just create a new column called salary_year_inflated in our dataframe df_salary
df_salary['salary_year_inflated']=df_salary['salary_year_avg'].apply(projected_salary)
#then show both the columns
#so well access the salary_year_inflated and salary_year_avg col from our df_salary dataframe
df_salary[['salary_year_avg','salary_year_inflated']]
```

```python
#even simpler code is using lambda
df_salary['salary_year_inflated']=df_salary['salary_year_avg'].apply(lambda salary: salary*1.03)
df_salary[['salary_year_avg','salary_year_inflated']]
```

```python
#now we need to convert this string to list
#we need to import the ast library
#so here we converted this rows datatype from string to list
import ast
ast.literal_eval(df['job_skills'][1])
```

```python
#now we converted this to a list
#so what we did first was converting this string to list
#now we will clean up this col using apply method
#now we will define a function that will convert job_skills col values from string to list
def clean_list(skill_list):
    #make sure that the value is not empty if not empty then only convert string to list
    if pd.notna(skill_list):
        return ast.literal_eval(skill_list)

#so above we defined a function that converts string to list
#now we will apply this function to all values of the job_skills col
#so im accessing the job_skills col from my dataframe and applying clean_list function on all values of that col to convert dtype from string to list
df['job_skills']=df['job_skills'].apply(clean_list)
```

```python
#so i created a new col salary_inflated by accessing the salary_year_avg and applying the apply() method and using lambda
#df_salary['salary_year_inflated']=df_salary['salary_year_avg'].apply(lambda salary: salary*1.03)
#df_salary[['salary_year_avg','salary_year_inflated']]
```

```python
#so in the above code we applied the apply method to a col by writing df['salary_year_avg']
#if we want to apply it on a row we dont have to mention the column just mention the df (only df.apply(method name,axis))
#first lets define the method and inside it we will be passing in the row
def projected_salary(row):
    #if the word senior is present in the row of the colum that we are checking(job_title_short)
    #then return the row of the salary_year_avg*1.05
    if "Senior" in row['job_title_short']:
        return 1.05*row['salary_year_avg']
    else:
        #if senior not present in the rows of the job_title_short col then return the row of the salary_year_avg but multiply by 1.03
        return 1.03*row['salary_year_avg']
    
#so now we craeted a new col salary_year_inflated of the df_salary by accessing the df_salary dataframe (as we want to apply the apply method on all the rows) and applying the apply(projected_salary) method to each rows
#we will apply the apply() method to the entire dataframe and not just to one particlar col
df_salary['salary_year_inflated']=df_salary.apply(projected_salary,axis=1)
df_salary[['job_title_short','salary_year_avg','salary_year_inflated']]
```
