```python
#import all neccessary libraries
import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset
import ast

#load the dataset
dataset=load_dataset('lukebarousse/data_jobs')
df=dataset['train'].to_pandas()

#data cleanup
df['job_posted_date']=pd.to_datetime(df['job_posted_date'])
df['job_skills']=df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)

#filter for data analyst data
#create a new df which shows only data analyst role
#create a copy of the original
df_DA=df[df['job_title_short']=='Data Analyst'].copy()
```

CUSTOMIZING OUR CHART FROM EXERCISE 11(EXERCISE_TRENDING_SKILLS.IPYNB)

```python
#created a new df that only shows data analyst jobs data
#create a copy of the original
df_DA=df[df['job_title_short']=='Data Analyst'].copy()

#create a new col for month no
df_DA['job_posted_month_no']=df_DA['job_posted_date'].dt.month 

#explode the job_skills col and store in new df
df_DA_explode=df_DA.explode('job_skills')

#create a pivot table which shows job_posted_month_no in row and job_skills in col
#agg func=size to count the no of times each skill appeared in each month
df_DA_pivot=df_DA_explode.pivot_table(index='job_posted_month_no',columns='job_skills',aggfunc='size',fill_value=0)

#now create a new row which provides the sum total of each col
#access row with the loc method
df_DA_pivot.loc['Total']=df_DA_pivot.sum()

#sort skills according to the count (total row)
df_DA_pivot=df_DA_pivot[df_DA_pivot.loc['Total'].sort_values(ascending=False).index]

#since we sorted we can drop the last row
df_DA_pivot=df_DA_pivot.drop('Total')

#use month names for plotting
df_DA_pivot=df_DA_pivot.reset_index()
#create a new col job_posted_month by accessing the job_posted_mont_no and applying the apply method

#lambda x: = "For each number x that you give me..."
#pd.to_datetime(x, format='%m') = "Treat this number as a month number and convert it to a date object"
#If x = 1  becomes "January 1st, 1900" (or some default date)
#If x = 12  becomes "December 1st, 1900"
#.strftime('%B') = "Now, from that date, extract just the full month name"
#"January 1st, 1900"  "January"
#"December 1st, 1900"  "December"

df_DA_pivot['job_posted_month']=df_DA_pivot['job_posted_month_no'].apply(lambda x: pd.to_datetime(x, format='%m').strftime('%B'))
df_DA_pivot=df_DA_pivot.set_index('job_posted_month')
df_DA_pivot=df_DA_pivot.drop(columns='job_posted_month_no')

#get the top 5 skills
df_DA_pivot.iloc[:,:5].plot(
    kind='line',
    linewidth=4,
    linestyle=':',
    colormap='viridis',
    marker='o',
    markersize=5,
    figsize=(10,5)
    )

plt.title('top 5 skills for a data analyst per month')
plt.ylabel('count')
plt.xlabel('')
plt.show()
```

CUSTOMIZING OUR SCATTER PLOT FROM 14_MATPLOTLIB_SCATTER_PLOT.IPYNB

```python
#explode the job_skills col into individual rows
df_exploded=df_DA.explode('job_skills')

#calculate the average salary and count of job postings per skill
skill_stats=df_exploded.groupby('job_skills').agg(
    #craete 2 col and apply the aggfunc to get it
    median_salary=('salary_year_avg','median'),
    skill_count=('job_skills','count')
)

#find only top 20 skills
skill_stats=skill_stats.sort_values(by='skill_count',ascending=False).head(20)
skill_stats
 
```

```python
skill_stats.plot(kind='scatter',x='skill_count',y='median_salary')


#now we dont know what these dots are representing
#we need to give proper names
#therefore loop through the skill_stats and only get the index(skill names not values (the counts))
#i=>index
#txt=> text that will be displayed
#enumerate=> gives index,value

for i,txt in enumerate(skill_stats.index):
    plt.text(skill_stats['skill_count'].iloc[i],skill_stats['median_salary'].iloc[i],txt)
    
plt.xlabel('count of job postings')
plt.ylabel('medain yearly salary')
plt.title('salary vs count of job postings for top 20 skills')
plt.tight_layout()
plt.show()
```

```python

#now we see that values are overlapping
#therefore do adjust text
#first start by importing this function

from adjustText import adjust_text

skill_stats.plot(kind='scatter',x='skill_count',y='median_salary')


#now we dont know what these dots are representing
#we need to give proper names
#therefore loop through the skill_stats and only get the index(skill names not values (the counts))
#i=>index
#txt=> text that will be displayed
#enumerate=> gives index,value

texts=[]
for i,txt in enumerate(skill_stats.index):
    texts.append(plt.text(skill_stats['skill_count'].iloc[i],skill_stats['median_salary'].iloc[i],txt))

adjust_text(texts, arrowprops=dict(arrowstyle="->",color='g',lw=0.5))

#get the current axis
ax=plt.gca()
#we want the y axis of this ax object (ax=axis)
#and we also want to modify the y axis
ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda y,pos: f'${int(y/1000)}K'))
  
plt.xlabel('count of job postings')
plt.ylabel('medain yearly salary')
plt.title('salary vs count of job postings for top 20 skills')
plt.tight_layout()
plt.show()
```
