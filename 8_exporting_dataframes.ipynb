```python
import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset

dataset=load_dataset('lukebarousse/data_jobs')
df=dataset['train'].to_pandas()

df['job_posted_date']=pd.to_datetime(df['job_posted_date'])

```

```python
df['job_posted_month']=df['job_posted_date'].dt.strftime('%b')
#created a list of months to store the unique months
months=df['job_posted_month'].unique()
months
```

```python
#build up the dictionary comprehension
{month:month for month in months}
```

```python
#but we want value to be dataframe of months
dict_months={month:df[df['job_posted_month']==month] for month in months}
dict_months['Jan']
```

```python
#now lets concatenate the df of jan feb and march(stack all the datas of jan,feb,march)
df_q1=pd.concat([dict_months['Jan'],dict_months['Feb'],dict_months['Mar']])
df_q1
```
